```python
s = 'назначить уголовное наказание в виде лишения свободы сроком 10 (десять) лет и 6 месяцев'

pattern = r"сроком*\s(.+)\s*(лет)*((.+)*месяц\w+)*"
result = re.search(pattern, s) 
result[0]
```

```python
def find_prison_term(text):
    return re.sub('сроком', '', re.search(r"сроком*\s(.+)\s*(лет)*((.+)*месяц\w+)*", text)[0]).strip()
s = 'назначить уголовное наказание в виде лишения свободы сроком 10 (десять) лет'
find_prison_term(s) 
```
