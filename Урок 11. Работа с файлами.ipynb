```python
with open('students.csv', encoding='Windows-1251') as file:  
    reader = csv.DictReader(file,delimiter = ';')
    for line in reader:
        if line['Наименование \nнаправления подготовки (специальности)'] == 'Военная журналистика':
            print(line['Число студентов, 2019'])
```

```python
columns=reader.fieldnames
```

```python
with open('students_2.csv', 'w', encoding='utf-8') as csv_file:
    file_writer = csv.DictWriter(csv_file, delimiter = ",", fieldnames=columns)
    file_writer.writeheader()
```

```python
file=open('students.csv', encoding='Windows-1251')
reader = csv.DictReader(file,delimiter = ';')

with open("students_2.csv", mode="w", encoding='utf-8') as csv_file:
    writer = csv.DictWriter(csv_file, delimiter = ",", fieldnames=columns)
    
    writer.writeheader()
    writer.writerows(reader)

file.close()
```
