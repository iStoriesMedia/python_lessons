### 1. Объединение датафреймов методами merge и concat

```python
df1 = pd.read_csv('/Users/sherlock/Downloads/IHME-GBD_2019_DATA-15798851-2/IHME-GBD_2019_DATA-15798851-2.csv')

# загружем 1-ый файл, сохранненный на компьютере
# ссылка на скачивание https://drive.google.com/file/d/1uK1edpdqVFJnFaGTWGpVOlaJHgoCBwli/view?usp=sharing 
# источник данных Global Burden of Disease Study 2019 (GBD 2019) Results.
# Seattle, United States: Institute for Health Metrics and Evaluation (IHME), 2020.
# Сайт http://ghdx.healthdata.org/gbd-results-tool
# DALYs (Disability-Adjusted Life Years) — этот показатель означает, 
# сколько лет здоровой и продуктивной жизни люди теряют из-за нетрудоспособности и ранней смерти от разных болезней

df1.head(10) # выводим первые 10 строк
```

```python
df2 = pd.read_csv("/Users/sherlock/Downloads/IHME-GBD_2019_DATA-ff08d9bc-1/IHME-GBD_2019_DATA-ff08d9bc-1.csv")

# загружем 1-ый файл, сохранненный на компьютере
# ссылка на скачивание https://drive.google.com/file/d/16fM_JEC5GAkYEDaDgzROrBm8loi9RZRZ/view
# источник данных Global Burden of Disease Study 2019 (GBD 2019) Results, сайт http://ghdx.healthdata.org/gbd-results-tool
# в файле данные о смертности людей от разных причин по 204 странам с 2000 по 2019 гг

df2.head(10)
```

```python
df1.rename(columns = {'val': 'daly_val', 'upper':'daly_upper', 'lower':'daly_lower'}, inplace = True)
df2.rename(columns = {'val': 'deaths_val', 'upper': 'deaths_upper', 'lower':'deaths_lower'}, inplace = True)
# меняем названия колонок в соответствии с показателями: число потерянных продуктивых лет для 1-го датафрейма (DALY)
# и смертность для 2-го датафрейма ()
```

```python
df_all = pd.merge(df1, df2, on = ['location', 'sex', 'age', 'cause', 'metric', 'year'], how = 'inner') 

# inner: берем только совпадения из двух датафреймов 
# у нас указаны названия колонок, по которым мы объединяем датафреймы, но также можно объединять по индексу — 
# порядковому номеру, который слева по вертикали c помощью [df1.index] без кавывек
```

```python
 # делим датафрейм на два датафрейма: 1-ый со строками от первой до 50 000, 2-ой от 50000 до последней
df_all_1 = df_all.iloc[:50000,:]
df_all_2 = df_all.iloc[50000:,:]
```

### 2.1. Добавление новых столбцов с помощью математических операций

```python
# добавляем новый столбец, в котором будет округленное значение из другого столбца
df_all["deaths_val_round"] = df_all["deaths_val"].round(decimals = 1) 
```

```python
df_all
```

```python
# добавляем новый столбец, суммируя значения нескольких других
df_all["deaths_total"] = df_all["deaths_val"] + df_all["deaths_upper"] + df_all["deaths_lower"]
df_all
```

```python
# добавляем новый столбец, суммируя значения нескольких других с помощью метода loc
# помимо sum, можно считать mean (среднее), median (медиана), max, min, std (стандарт.отклонение) и пр.
df_all["total_loc"] = df_all.loc[:, "deaths_val":"deaths_lower"].sum(axis = 1)
df_all
```

```python
# добавляем новый столбец, суммируя значения нескольких других с помощью метода iloc
# важно, что при использовании метода iloc правое значение (номер столбца) не включается, поэтому если нам нужны столбцы с номерами
# 11, 12 и 13, указать нужно по с 11 по 14-й: df_all.iloc[:, 11:14] 

df_all["total_iloc"] = df_all.iloc[:, 11:14].sum(axis = 1)
```

```python
df_all
```

```python
# удалим ненужные столбцы, чтобы не мешали
df_all.drop(['deaths_val_round','deaths_total', 'total_loc','total_iloc'], axis = 1, inplace = True)
```

### 2.2. Добавление новых столбцов с помощью lambda функций

```python
# создаем список стран, входивших в 2020 году в СНГ
CIS_2020 = ['Azerbaijan', 'Armenia', 'Belarus', 'Kazakhstan', 'Kyrgyzstan', 
            'Moldova', 'Russian Federation', 'Tajikistan', 'Uzbekistan']
```

```python
# с помощью lambda-функции прописываем, что если страна ('location') входит в СНГ, то пусть в новом столбце 'CIS_2020' 
# будет True, а если нет, то False
df_all['CIS_2020'] = df_all['location'].apply(lambda x: True if x in CIS_2020 else False )
df_all[df_all['CIS_2020'] == True]
```

```python
# можно записать и короче, не прописывая полностью это условие
df_all['CIS_2020'] = df_all['location'].apply(lambda x: x in CIS_2020)
```

```python
# выберем строки, которые соответствуют новому условию и проверим, удалось ли их добавить 
df_all[df_all['CIS_2020'] == True]
```

```python
# точно так же можно прописать любое сложное условие отдельной функцией, а затем применить к нужному столбцу
def if_cis(x):
    if x in CIS_2020:
        return True
    else:
        False
```

```python
df_all['CIS_2020'] = df_all['location'].apply(if_cis)
```

```python
df_all[df_all['CIS_2020'] == True]
```

```python
# сохраняем датафрейм в csv
df_all.to_csv('df_all.csv')
```

### 3. Работа с большими файлами с помощью аргумента chunksize

```python
# разобьем большой файл на части с помощью аргумента chunksize, в котором прописывается сколько строк должно быть в каждой части
# выведем по 5 строк каждой получившейся части

for chunk_df in pd.read_csv('df_all.csv', 
                      chunksize = 50000):
    print("CHUNK DF")
    print(chunk_df.head())
```

```python
# применим метод groupby отдельно к каждой части и затем объединим результат в новый датафрейм
new_df = pd.DataFrame() # пустой датафрейм
for chunk_df in pd.read_csv('df_all.csv', 
                      chunksize = 50000):
    result = chunk_df.groupby(['cause', 'year']).agg({'deaths_val': 'mean',
                                                    'daly_val': 'mean'})
    new_df = pd.concat([new_df,result])
```

```python
new_df
```

```python
# сохраним результат в файл
new_df.to_csv('deaths_daly_groupby.csv', encoding = 'utf-8')
```
